```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def min_filter(image, kernel_size=3):
    # Apply Min filter using the erode function
    min_filtered = cv2.erode(image, np.ones((kernel_size, kernel_size), np.uint8))
    return min_filtered

# Load the image
image = cv2.imread('/content/pika.jpg.png', cv2.IMREAD_GRAYSCALE)

# Check if image is loaded successfully
if image is None:
    print("Error loading image")
else:
    # Apply the Min filter with the desired kernel size
    filtered_image = min_filter(image, kernel_size=5)

    # Display the original and filtered images
    plt.figure(figsize=(10, 5))

    # Original Image
    plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')

    # Min Filtered Image
    plt.subplot(1, 2, 2), plt.imshow(filtered_image, cmap='gray'), plt.title('Min Filtered Image')

    plt.show()
```
