```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def median_filter(image, kernel_size=3):
    # Apply Median filter using the medianBlur function
    median_filtered = cv2.medianBlur(image, kernel_size)
    return median_filtered

# Load the image
image = cv2.imread('/content/images.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if image is None:
    print("Error loading image")
else:
    # Apply the Median filter with the desired kernel size
    filtered_image = median_filter(image, kernel_size=5)

    # Display the original and filtered images
    plt.figure(figsize=(10, 5))

    # Original Image
    plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')

    # Median Filtered Image
    plt.subplot(1, 2, 2), plt.imshow(filtered_image, cmap='gray'), plt.title('Median Filtered Image')

    plt.show()
```
