```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

def max_filter(image, kernel_size=3):
    # Apply Max filter using the dilate function
    max_filtered = cv2.dilate(image, np.ones((kernel_size, kernel_size), np.uint8))
    return max_filtered

# Load the image
image = cv2.imread('/content/face_recognition.jpg', cv2.IMREAD_GRAYSCALE)

# Check if the image is loaded successfully
if image is None:
    print("Error loading image")
else:
    # Apply the Max filter with the desired kernel size
    filtered_image = max_filter(image, kernel_size=5)

    # Display the original and filtered images
    plt.figure(figsize=(10, 5))

    # Original Image
    plt.subplot(1, 2, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')

    # Max Filtered Image
    plt.subplot(1, 2, 2), plt.imshow(filtered_image, cmap='gray'), plt.title('Max Filtered Image')

    plt.show()
```
